```python
import pandas as pd
from pandasql import sqldf
```

```python
races = pd.read_csv("./raw/races.csv")
contructors = pd.read_json("./raw/constructors.json", lines=True)
drivers = pd.read_json("./raw/drivers.json", lines=True)
results = pd.read_json("./raw/results.json", lines=True)
```

## Transform data

```python
races.rename(columns={'year': 'raceYear'}, inplace=True)
```

```python
contructors.rename(columns={"name": "teamName"}, inplace=True)
```

```python
drivers["driverName"] = drivers["name"].apply(lambda x: x["forename"] + " " + x["surname"])
del drivers["name"]
```

```python
result_driver = results.merge(drivers, on="driverId", how="inner")
resut_driver_constructor = result_driver.merge(contructors, on="constructorId", how="inner")
calculated_race_result = resut_driver_constructor.merge(races, on="raceId", how="inner")
```

```python
calculated_race_result.columns
```

```python
calculated_race_result = calculated_race_result[["raceYear", "teamName", "driverName", "position", "points"]]
```

```python
def convert_position_from_str_to_int(inp):
    try:
        return int(inp)
    except Exception as e:
        return 999
calculated_race_result["int_position"] = calculated_race_result["position"].apply(convert_position_from_str_to_int)
```

```python
top_calculated_race_result = calculated_race_result[calculated_race_result["int_position"] < 10]
del top_calculated_race_result["position"]
top_calculated_race_result.rename(columns={"int_position": "position"}, inplace=True)
```

```python
top_calculated_race_result["calculatedPoints"] = 11 - top_calculated_race_result["position"]
```

```python
top_calculated_race_result.head(5)
```

```python
top_calculated_race_result.rename(columns={"raceYear": "race_year", "teamName": "team_name", "driverName": "driver_name", "calculatedPoints": "calculated_points"}, inplace=True)
```

## Analysis

```python
dominant_teams = sqldf("""
SELECT team_name,
       COUNT(1) AS total_races,
       SUM(calculated_points) AS total_points,
       AVG(calculated_points) AS avg_points,
       RANK() OVER(ORDER BY AVG(calculated_points) DESC) team_rank
  FROM top_calculated_race_result
GROUP BY team_name
HAVING COUNT(1) >= 100
ORDER BY avg_points DESC
""")
```

```python
result_dominant_teams = sqldf("""
SELECT race_year, 
       team_name,
       COUNT(1) AS total_races,
       SUM(calculated_points) AS total_points,
       AVG(calculated_points) AS avg_points
  FROM top_calculated_race_result
 WHERE team_name IN (SELECT team_name FROM dominant_teams WHERE team_rank <= 5)
GROUP BY race_year, team_name
ORDER BY race_year, avg_points DESC
""")
```

```python
result_dominant_teams.head(5)
```

```python
dominant_drivers = sqldf("""
SELECT driver_name,
       COUNT(1) AS total_races,
       SUM(calculated_points) AS total_points,
       AVG(calculated_points) AS avg_points,
       RANK() OVER(ORDER BY AVG(calculated_points) DESC) driver_rank
  FROM top_calculated_race_result
GROUP BY driver_name
HAVING COUNT(1) >= 50
ORDER BY avg_points DESC
""")
```

```python
result_dominant_drivers = sqldf("""
SELECT race_year, 
       driver_name,
       COUNT(1) AS total_races,
       SUM(calculated_points) AS total_points,
       AVG(calculated_points) AS avg_points
  FROM top_calculated_race_result
 WHERE driver_name IN (SELECT driver_name FROM dominant_drivers WHERE driver_rank <= 10)
GROUP BY race_year, driver_name
ORDER BY race_year, avg_points DESC
""")
```

```python
result_dominant_drivers.head(5)
```

```python
set(result_dominant_drivers["driver_name"])
```

## Visualization

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

# Create plots for each driver
for driver, group in result_dominant_drivers.groupby('driver_name'):
    ax.plot(group['race_year'], group['avg_points'], label=driver, marker='', linestyle='-')

# Add labels and title
ax.set_xlabel('Race Year')
ax.set_ylabel('Average Points')
ax.set_title('Average Points per Race by Year')

# Move legend outside the chart
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show plot
plt.show()
```

```python
pd.set_option('display.max_rows', 10)
```

```python
grouped_df = result_dominant_drivers.groupby('driver_name').sum().reset_index()

# Create a bar chart
fig, ax = plt.subplots()

bar_width = 0.35
index = grouped_df.index
opacity = 0.8

bar1 = ax.bar(index, grouped_df['total_races'], bar_width, alpha=opacity, label='Total Races')
bar2 = ax.bar(index + bar_width, grouped_df['total_points'], bar_width, alpha=opacity, label='Total Points')

# Add labels and title
ax.set_xlabel('Driver Name')
ax.set_ylabel('Values')
ax.set_title('Total Races and Total Points by Driver')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(grouped_df['driver_name'], rotation=90)  # Rotate labels vertically
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```

## Top Performance Formula 1 Drivers (1965-2020)

This dataset presents the top-performing Formula 1 drivers each year, along with their total races, total points earned, and average points per race.

#### Key Insights:

- **Driver Representation:** The dataset showcases the dominant drivers in Formula 1 racing across several decades, including iconic names like Jackie Stewart, Niki Lauda, Alain Prost, Nigel Mansell, Ayrton Senna, Michael Schumacher, and Lewis Hamilton.
Performance Metrics:
- **Total Races:** The number of races each driver participated in during the respective year.
- **Total Points:** Cumulative points earned by the driver throughout the season.
- **Average Points:** The average points scored per race, providing a measure of consistency and performance level.
    
#### Notable Performances:

- **Jackie Stewart:** Demonstrated consistent high performance throughout the late 1960s and early 1970s, achieving impressive average points per race.
- **Niki Lauda, Alain Prost, and Michael Schumacher:** Dominated different eras with multiple top performances and high average points per race.
- **Lewis Hamilton:** Emerged as a dominant force in recent years, consistently achieving high average points and leading in total points earned.

#### Insights:

Analyzing the top-performing drivers and their average points per race offers valuable insights into their consistency and competitiveness throughout their careers, contributing to the rich history and legacy of Formula 1 racing.
